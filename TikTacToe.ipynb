```python

def display(turn, positions_record, player1, player2):

    
    # Board representation
    
    line1={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line2={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line3={"1":'_____',"2":'|',"3":'_____',"4":'|',"5":'_____'}
    line4={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line5={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line6={"1":'_____',"2":'|',"3":'_____',"4":'|',"5":'_____'}
    line7={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line8={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}
    line9={"1":'     ',"2":'|',"3":'     ',"4":'|',"5":'     '}



    
    ###########################################################################################################################
    
    # Determine which player is making the move based on the turn
    
    if turn % 2 == 0:
        current_player = player2
    else:
        current_player = player1

    ###########################################################################################################################
    
    
    #Update the board based on the player's move (position)
            
            #pos_1 = line2[1]
            #pos_2 = line2[3]
            #pos_3 = line2[5]
            #pos_4 = line5[1]
            #pos_5 = line5[3]
            #pos_6 = line5[5]
            #pos_7 = line8[1]
            #pos_8 = line8[3]
            #pos_9 = line8[5]
    
    
    for position, space in positions_record.items():
        
        if space == '':
            
            pass
        
        else:
            
            if position == '1':
                line2['1'] = positions_record['1']
        
            elif position == '2':
                line2['3'] = positions_record['2']
            
            elif position == '3':
                line2['5'] = positions_record['3']
        
            elif position == '4':
                line5['1'] = positions_record['4']
             
            elif position == '5':
                line5['3'] = positions_record['5']
        
            elif position == '6':
                line5['5'] = positions_record['6']
        
            elif position == '7':
                line8['1'] = positions_record['7']
             
            elif position == '8':
                line8['3'] = positions_record['8']
        
            elif position == '9':
                line8['5'] = positions_record['9']
        

    ###########################################################################################################################
    
    # Print the board
    
    print(line1["1"],line1["2"],line1["3"],line1["4"],line1["5"])
    print(line2["1"],line2["2"],line2["3"],line2["4"],line2["5"])
    print(line3["1"],line3["2"],line3["3"],line3["4"],line3["5"])
    print(line4["1"],line4["2"],line4["3"],line4["4"],line4["5"])
    print(line5["1"],line5["2"],line5["3"],line5["4"],line5["5"])
    print(line6["1"],line6["2"],line6["3"],line6["4"],line6["5"])
    print(line7["1"],line7["2"],line7["3"],line7["4"],line7["5"])
    print(line8["1"],line8["2"],line8["3"],line8["4"],line8["5"])
    print(line9["1"],line9["2"],line9["3"],line9["4"],line9["5"])     
    
############################################################################################################################################
############################################################################################################################################
############################################################################################################################################

def check_3_in_row(positions_record, player1, player2):
    
    # From the symbol determine the player 
    symbol_to_player = {player1: "Player1", player2: "Player2"}

    # Check all winning combinations
    winning_combinations = [ ('1', '2', '3'), ('4', '5', '6'),('7', '8', '9'),('1', '4', '7'),
                            ('2', '5', '8'),('3', '6', '9'),('1', '5', '9'),('3', '5', '7')]

    for combination in winning_combinations:
        if positions_record[combination[0]] == positions_record[combination[1]] == positions_record[combination[2]] and positions_record[combination[0]] != '':
            
            winner_position = positions_record[combination[0]]  
            winner_symbol =  winner_position[2]
            
            return f'{symbol_to_player[winner_symbol]} wins!'

    return 'No winner yet'
         

############################################################################################################################################
############################################################################################################################################
############################################################################################################################################

def player(turn,player1,player2):

# Set the current player
    
    if turn % 2 == 0:
        current_player = player2
    else:
        current_player = player1

    return current_player
    

############################################################################################################################################
############################################################################################################################################
############################################################################################################################################

def tictactoe():

    
    # An empty board is shown
    
    turn = 0

    positions_record = {'1': '' ,'2': '' ,'3': '' ,'4': '' ,'5': '' ,'6': '' ,'7': '' ,'8': '' ,'9': '' }
      
    player1 = 'X'

    player2 = 'O'

    display(turn, positions_record, player1, player2)
        
    
    ###########################################################################################################################
    
    
    # Ask player 1 for their choice (X or O)

    while player1 not in ['x', 'o', 'the x', 'the o']:          # Continues to ask until choice = X or O
        
        player1 = input("Do you want to play a game of Tic Tac Toe? Player 1, what will you be? The X or the O? ").lower()
        

    
    ###########################################################################################################################
    
    
    # Assign players to X and O
    
    if player1 in ['x', 'the x']:
        player1 = 'X'
        player2 = 'O'
    else:
        player1 = 'O'
        player2 = 'X'
   
    
    ###########################################################################################################################

    
    # First turn : player 1 choses its play position 
    
    
    turn += 1
    
    played_positions = []

    position = ''
 
    while not position.isdigit() or int(position) not in range(1, 10):
        position = input("Great! Lets begin! Where are you going to play? Choose a position from 1-9 : ")
               

    current_player = player(turn,player1,player2)
    
    positions_record[position] = f'  {current_player}  '

    played_positions.append(position)

    position = ''

    turn += 1

    ###########################################################################################################################
    
    
    #Shows the board after the position is picked
    
          
    display(turn, positions_record, player1, player2)
    


    ###########################################################################################################################

    
    # Next turns: 

    player1_or_2 = 'Player2'
   
    while check_3_in_row(positions_record, player1, player2) == 'No winner yet':

        while not position.isdigit() or int(position) not in range(1, 10) or position in played_positions:
            
            position = input(f"{player1_or_2}, it's your turn! Choose a position from 1 to 9 : ")

               
        current_player = player(turn,player1,player2)
        
        positions_record[position] = f'  {current_player}  '

        played_positions.append(position)

        position = ''

        turn += 1
        
        if turn %2 == 0:
            player1_or_2 = 'Player2'
        else:
            player1_or_2 = 'Player1'

        
        display(turn, positions_record, player1, player2)

        
 
    return check_3_in_row(positions_record, player1, player2)
        
        
        

        

        
    
    

    
    
```
